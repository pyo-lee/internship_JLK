```python
import os, glob, cv2, math
import numpy as np
import matplotlib.pyplot as plt

def imshow_plt(img, show_type='color'):
    plt.figure(figsize=(10,10))
    if show_type=='gray':
        plt.imshow(img, 'gray')
    else:
        plt.imshow(img)
    plt.show()
    
```

```python
PATH = '../03_preprocess/crop_re/'
label_folder = os.listdir(PATH)
label_li = [float(l.split('_')[-1]) for l in label_folder if not '.' in l]
label_folder.sort()
print('*'*30)
print('Load data')
print('*'*30)

for label_path in label_folder:
    print(label_path)
    if not '.' in label_path:
        img_path_tmp = glob.glob(PATH+label_path+'/*')
    #     images=list()
        if not os.path.isdir('../03_preprocess/crop_wb/'+label_path):
                os.makedirs('../03_preprocess/crop_wb/'+label_path)
                
        for path in img_path_tmp:
            img = cv2.imread(path)
            img_mean = (img*1.0 / img.mean(axis=(0,1)))
            cv2.imwrite(path.replace('crop_re', 'crop_wb'), img_mean*255)
```
